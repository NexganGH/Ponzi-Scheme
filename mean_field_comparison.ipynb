```python
from networks import *
from simulation import *
```

```python
er = ErNetwork.load_json('my_networks/er1.json')

par = PonziParameters(M=1, starting_capital=50, lambda_=lambda _: 0.2, mu=lambda _: 0.1, rp=lambda _:0.2, rr=lambda _: 0.1)
sim = PonziSimulation(er, par, 30 * 12, 1./12)
results = sim.simulate_ponzi(computed_for_each_k=True)
```

```python
plot = PonziPlotter()
for k in [3, 6, 9]:
    res = results[k]
    plot.add_simulation(res, label='k={}'.format(k))
plot.plot(show_investor=True, show_potential=True)
```
