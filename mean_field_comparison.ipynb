```python
from networks import *
from simulation import *
```

```python
er = ErNetwork.load_json('my_networks/er1.json')

par = PonziParameters(M=1, starting_capital=50, lambda_=lambda _: 0.2, mu=lambda _: 0.1, rp=lambda _:0.2, rr=lambda _: 0.1)
sim = PonziSimulation(er, par, 30 * 12, 1./12)
results_er = sim.simulate_ponzi(computed_for_each_k=True)
```

```python
plot = PonziPlotter()
for k in [0, 3, 6, 10]:
    res = results_er[k]
    lab = 'Sim. - k={}'.format(k) if k != 0 else "Sim. - Totale"

    plot.add_simulation(res, label=lab)
plot.plot(show_investor=True, show_potential=True, title='Simulazione su Rete ER', file_name='meanfield/meanfield_sim_er')
```
