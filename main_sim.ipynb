```python
lambdas_to_test = [0.05, 0.075, 0.1, 0.125, 0.150, 0.175, 0.2]
```

```python

plotter = PonziPlotter()
#plotter2 = PonziPlotter()

par_list = []
for x in lambdas_to_test:
    erNet = ErNetwork.load_json('my_networks/er1.json')
    par = PonziParameters(M=1, starting_capital=50, lambda_ = lambda t: x, mu=lambda t:0.05, rp= lambda t: 0.1, rr=lambda t: 0.05)
    sim = PonziSimulation(erNet, par, max_time_units=30*12)
    res = sim.simulate_ponzi()
    plotter.add_simulation(res, r'$\lambda=' + str(x) + '$')
plotter.plot('changing_lambda_er_investors', show_investor=True, show_deinvestor=True)
plotter.plot('changing_lambda_er_capital', show_capital=True)
```

```python
plotter = PonziPlotter()

for x in lambdas_to_test:
    par = PonziParameters(M=1, starting_capital=50, lambda_ = lambda t: x, mu=lambda t:0.05, rp= lambda t: 0.1, rr=lambda t: 0.05)
    equations = DifferentialEquations(N=10000, ponzi_parameters=par, avg_k=6)
    res = equations.solve(0, 30, 1000)
    plotter.add_simulation(res, r'$\lambda=' + str(x) + '$')

plotter.plot('changing_lambda_eq', show_investor=True, show_deinvestor=True)
plotter.plot('changing_lambda_eq_capital', show_capital=True)
```

```python
erNet, baNet = ErNetwork.load_json('my_networks/er1.json'), BaNetwork.load_json('my_networks/ba1.json')
wsNet = WattsStrogatzNetwork.load_json('my_networks/ws1.json')
networks = [erNet, baNet, wsNet]
par = PonziParameters(M=1, starting_capital=50, lambda_ = lambda t: 0.01, mu=lambda t:0.1, rp= lambda t: 0.1, rr=lambda t: 0.05)

labels = ['ER', 'BA', 'WS']
i=0
plotter = PonziPlotter()
for (net, lab) in zip(networks, labels):
    sim = PonziSimulation(net, par, max_time_units=30*12)
    res = sim.simulate_ponzi()
    plotter.add_simulation(res, lab)
plotter.plot('confronto_lambda_crit', show_investor=True, show_capital=True)
```
