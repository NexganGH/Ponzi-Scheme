```python

plotter = PonziPlotter()
#plotter2 = PonziPlotter()

par_list = []
for x in lambdas_to_test:
    erNet = ErNetwork.load_json('my_networks/er1.json')
    par = PonziParameters(M=1, starting_capital=50, lambda_ = lambda t: x, mu=lambda t:mu, rp= lambda t: rp, rr=lambda t: 0.05)
    sim = PonziSimulation(erNet, par, max_time_units=30*12)
    res = sim.simulate_ponzi()
    plotter.add_simulation(res, r'$\lambda=' + str(x) + '$')
plotter.plot('changing_lambda/changing_lambda_er_investors', show_potential=True, show_investor=True, show_deinvestor=True, title="Popolazione Rete ER")
plotter.plot('changing_lambda/changing_lambda_er_capital', show_capital=True, title="Capitale Rete ER")
```

```python
plotter = PonziPlotter()

for x in lambdas_to_test:
    par = PonziParameters(M=1, starting_capital=50, lambda_ = lambda t: x, mu=lambda t:mu, rp= lambda t: 0.1, rr=lambda t: 0.05)
    equations = DifferentialEquations(N=10000, ponzi_parameters=par, avg_k=6)
    res = equations.solve(0, 30, 1000)
    plotter.add_simulation(res, r'$\lambda=' + str(x) + '$')

plotter.plot('changing_lambda/changing_lambda_eq_investors', show_potential=True, show_investor=True, show_deinvestor=True, title="Popolazione Eq. Campo Medio Con Runge-Kutta")
plotter.plot('changing_lambda/changing_lambda_eq_capital', show_capital=True, title="Capitale Eq. Campo Medio Con Runge-Kutta")
```

```python
lambda_ = 0.2
k = 6
tau = 1/(k * lambda_ - mu)
import numpy as np
def i_approx(t):
    return 1/10000 / ((1 + 1/10000 * np.exp(t/tau)))**(lambda_ * k * tau) * np.exp(t/tau)

t_space = np.linspace(0, 30, 1000)
import matplotlib.pyplot as plt
plt.plot(t_space, i_approx(t_space))
```
