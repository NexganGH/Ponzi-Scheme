```python
from collections import Counter
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from networks import Network


# Get the degree distribution
degrees = [node.k() for node in net.nodes]
degree_counts = Counter(degrees)
k_values, frequencies = zip(*sorted(degree_counts.items()))

# Define the power-law function
def power_law(k, gamma, a):
    return a * k**(-gamma)

# Fit the power-law
k_values = np.array(k_values)
frequencies = np.array(frequencies)
fit_indices = k_values > 2  # Exclude k=1 and k=2
params, _ = curve_fit(power_law, k_values[fit_indices], frequencies[fit_indices])
gamma, a = params

# Plot
plt.loglog(k_values, frequencies, 'bo', label='Data')
plt.loglog(k_values, power_law(k_values, gamma, a), 'r-', label=f'Fit (gamma={gamma:.2f})')
plt.xlabel('k')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
