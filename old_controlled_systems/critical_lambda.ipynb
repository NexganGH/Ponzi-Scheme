```python
networks_to_test = [('eq', 'Equazioni Differenziali'), (lambda: ErNetwork.load_json('../my_networks/er1.json'), 'Network ER'), (lambda: BaNetwork.load_json(
    '../my_networks/ba1.json'), 'Network BA'), (lambda: WsNetwork.load_json('../my_networks/ws1.json'), 'Network WS')]
```

```python
i = 0
for create_network, lbl in networks_to_test:
    plotter = PonziPlotter()
    for lambda_ in [0.025, 0.05, 0.1]:
        par = PonziParameters(M=1, starting_capital=50, lambda_ = lambda t: lambda_, mu=lambda t:0.1, rp= lambda t: 0.1, rr=lambda t: 0.05)
        if create_network == 'eq':
            equations = DifferentialEquations(N=10000, ponzi_parameters=par, avg_k=6)
            res = equations.solve(0, 30, 1000)
        else:
            net = create_network() # carica la rete
            sim = PonziSimulation(net, par, max_time_units=30 * 12)
            res = sim.simulate_ponzi()
        plotter.add_simulation(res, r'$\lambda=' + str(lambda_) + '$')
    plotter.plot(f'critical/critical_{i}_investors', lbl, show_investor=True, show_deinvestor=True)
    plotter.plot(f'critical/critical_{i}_capital', lbl, show_capital=True, max_capital=13000)
    i+= 1
```
